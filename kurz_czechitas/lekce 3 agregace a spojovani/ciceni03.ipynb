```python
# Načteme tabulky o potravinách
import pandas as pd
url_food_sample_100 = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/nacteni-dat/nacteni-dat/food_sample_100.csv"
url_food_other = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/spojovani/sjednoceni-dat/food_other.csv"
url_food_nutrient = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/podmineny-vyber/podmineny-vyber/food_nutrient.csv"
url_branded_food = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/spojovani/excs/znacky/branded_food.csv"
food_sample_100 = pd.read_csv(url_food_sample_100)
food_other = pd.read_csv(url_food_other)
food_nutrient = pd.read_csv(url_food_nutrient)
branded_food = pd.read_csv(url_branded_food)
food = pd.concat([food_sample_100, food_other], ignore_index=True)
food_merge = pd.merge(food,food_nutrient, on ='fdc_id', how = 'inner') # HOW = 'left'/'right'/'inner'/'outer'/'cross'/
food_merged_brands = pd.merge(food_merge,branded_food, on='fdc_id', how='left') 
food_merged_brands = food_merged_brands.rename(columns={"name": "nutrition_name"})
```

```python
food_merged_brands_lipid = food_merged_brands[food_merged_brands["nutrition_name"] == "Total lipid (fat)"]
food_merged_brands_lipid.head(3)
```

```python
# proveď agregaci podle názvu kategorie s tím, že vypočítej průměrné množství tuků. Výslednou tabulku seřaď sestupně
lipid_agg = food_merged_brands_lipid.groupby('branded_food_category')['amount'].mean().sort_values(ascending=False)
lipid_agg
```

```python
food_merged_brands.groupby(by=['branded_food_category','brand_owner'])['fdc_id'].nunique().sort_values(ascending=False)
```

```python
# hlas volby
import pandas as pd
votes = pd.read_csv("votes.csv")
votes.head(5)
```

```python
#Celkový počet odevzdaných hlasů v jednotlivých letech
votes.groupby("year")["totalvotes"].sum().sort_values(ascending=False)
```

```python
#Celkový počet odevzdaných hlasů pro kandidáty a kandidátky jednotlivých stran pro volby v roce 2020.
votes2020 = votes[votes['year']==2020]
votes2020.groupby(by=['party_detailed','candidate'])['candidatevotes'].sum().sort_values(ascending=False)
```

```python
#Celkový počet odevzdaných hlasů pro kandidáty dvou hlavních stran (DEMOCRAT a REPUBLICAN) pro všechny volby od roku 2020.
year = 2020
filter = (votes['party_detailed'].isin(['DEMOCRAT','REPUBLICAN'])) &  (votes['year']>=year)
votes[filter].groupby(by=['year','party_simplified'])['candidatevotes'].sum().sort_values(ascending=False)
```

```python
#Celkový počet odevzdaných hlasů pro stát MONTANA pro všechny roky.
votesMontana = votes[votes['state']=='MONTANA']
votesMontana.groupby('state')['candidatevotes'].sum()
```

```python
#pořadí pro každý stát a rok aby číslo 1 měl(a) vždy vítěz(ka) 
votes['rank'] = votes[filter].groupby(by=['state','year'])['candidatevotes'].rank(ascending=False)
#Vytvoř tabulku, která bude obsahovat vítěznou stranu v jednotlivých státech ve volbách v roce 2020
podminka1 = votes['year']==2020
podminka2 = votes['rank']==1
winners_by_state_2020 = votes[(podminka1) &(podminka2)][['year','state','party_detailed']]
winners_by_state_2020.head()
```
