```python
import pandas as pd
penguins_adelie = pd.read_csv("penguins_adelie.csv")
penguins_chinstrap = pd.read_csv("penguins_chinstrap.csv")
penguins_gentoo = pd.read_csv("penguins_gentoo.csv")
```

```python
penguins_all = pd.concat([penguins_adelie,penguins_chinstrap,penguins_gentoo])
penguins_all.shape[0]
```

```python
penguins_all.head(2)
```

```python
island_data = {
    'Island': ['Torgersen', 'Biscoe', 'Dream'],
    'Latitude': [-64.7667, -66.0000, -64.7333],
    'Longitude': [-64.0667, -66.0000, -64.2333],
    'Location': [
        'Torgersen Island, near Palmer Station on the Antarctic Peninsula, close to Anvers Island',
        'Biscoe Islands, an archipelago off the west coast of the Antarctic Peninsula, south of the Palmer Archipelago',
        'Dream Island, part of the Argentine Islands, located near the western side of the Antarctic Peninsula'
    ]
} 
penguin_islands_df = pd.DataFrame(island_data)
penguin_islands_df.head(2)
```

```python
penguins_all_merged = pd.merge(penguins_all,penguin_islands_df, on='Island') 
penguins_all_merged.describe()
```

```python
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
penguins = pd.read_csv(url)
penguins_isna = penguins.isna()
penguins_isna.groupby('year').sum()
```

```python
penguins.head()
```

```python
penguins.groupby('species')['body_mass_g'].mean()
```

```python
penguins.groupby('species')['body_mass_g'].max().sort_values(ascending=False)
```

```python
penguins.groupby(['species','sex']).size().sort_values(ascending=False)
```

```python
penguins_min_flipper = penguins.groupby(['species'])['flipper_length_mm'].min()
penguins_max_flipper = penguins.groupby(['species'])['flipper_length_mm'].max()
penguins_flippe = pd.merge(penguins_min_flipper,penguins_max_flipper, on='species') 
penguins_flippe['diff'] = penguins_flippe['flipper_length_mm_y']-penguins_flippe['flipper_length_mm_x']
penguins_flippe
```

```python
penguins.groupby('island')['body_mass_g'].median().sort_values(ascending=False)
```
