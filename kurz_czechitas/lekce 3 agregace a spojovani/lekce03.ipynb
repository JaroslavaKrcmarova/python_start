# 3.lekce AGREGACE A SPOJOVANI

```python
# nacteni dat
import pandas as pd
url_food_sample_100 = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/nacteni-dat/nacteni-dat/food_sample_100.csv"
url_food_other = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/spojovani/sjednoceni-dat/food_other.csv"
url_food_nutrient = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/podmineny-vyber/podmineny-vyber/food_nutrient.csv"
url_branded_food = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/spojovani/excs/znacky/branded_food.csv"
food_sample_100 = pd.read_csv(url_food_sample_100)
food_other = pd.read_csv(url_food_other)
food_nutrient = pd.read_csv(url_food_nutrient)
branded_food = pd.read_csv(url_branded_food)
```

```python
# nahled
food_sample_100
```

```python
# velikost dat
print(food_sample_100.shape)
print(food_other.shape)
```

```python
# spojeni tabuli - CONCAT() = SQL UNION - spoji i nestejne ale atributy z jedne tab ktere nejsou v druhe naplni null hodnotami
food = pd.concat([food_sample_100,food_other])
food
```

```python
#reset index
food = pd.concat([food_sample_100, food_other], ignore_index=True)
food
```

```python
food_nutrient
```

```python
# spojeni pres klic = SQL JOIN
pd.merge(food,food_nutrient)    #spoji pres stejne nazvany klic
pd.merge(food,food_nutrient, on ='fdc_id')    # ON explicitne vybrany klic #default je inner join
food_merge = pd.merge(food,food_nutrient, on ='fdc_id', how = 'inner') # HOW = 'left'/'right'/'inner'/'outer'/'cross'/
food_merge
```

```python
#cviceni
branded_food.head(5)
```

```python
#cviceni Zjisti tři výrobce s největším počtem potravin v tabulce.
branded_food["brand_owner"].value_counts().head(3)
```

```python
#cviceni Zjisti pět kategorií s největších počtem potravin v tabulce.
branded_food["branded_food_category"].value_counts().head(3)
```

```python
print(branded_food.shape)
food_merge = pd.merge(food,food_nutrient, on ='fdc_id', how = 'inner') # HOW = 'left'/'right'/'inner'/'outer'/'cross'/
food_merge.shape
```

```python
# cvicenie inner join 
food_merged_brands = pd.merge(food_merge,branded_food, left_on='fdc_id', right_on='fdc_id') 
food_merged_brands.shape
```

```python
# cvicenie left join 
food_merged_brands = pd.merge(food_merge,branded_food, on='fdc_id', how='left') 
food_merged_brands.shape
```

```python
food_merged_brands.columns
```

```python
food_merged_brands = food_merged_brands.rename(columns={"name": "nutrition_name"})
food_merged_brands.columns
```

```python
food_merged_brands.head()
```

```python
#filtrujeme protein a prumerny mnozstvi
#protein = food_merged_brands[food_merged_brands["nutrient_name"] == "Protein"]
protein = food_merged_brands[food_merged_brands["nutrition_name"] == "Protein"]
```

```python
#prumer pres kategorii
protein.groupby('branded_food_category')['amount'].mean()
```

```python
# serazeni 
protein.groupby('branded_food_category')['amount'].mean().sort_values(ascending=False)
```

```python
protein.groupby("branded_food_category")["amount"].mean().sort_values(ascending=False)
```
