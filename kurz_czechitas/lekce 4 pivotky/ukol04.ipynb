```python
import pandas as pd
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
data = pd.read_csv(url)
data.head()
```

```python
bins = [-1,180,200,float('inf')]
labels = ['short','medium','long']
data['category'] = pd.cut(data['flipper_length_mm'],bins= bins,labels=labels)
data[data['category'] =='long']
```

```python
pd.crosstab(index=data['species'], columns=data['island'])
```

```python
pd.crosstab(index=data['species'], columns=data['island'],normalize='index')
```

```python
pd.pivot_table(data,
               index=data['species'],
               columns=data['sex'],
               aggfunc={'bill_depth_mm': "mean"}
               )
```
