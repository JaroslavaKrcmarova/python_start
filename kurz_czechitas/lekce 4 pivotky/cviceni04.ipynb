```python
import pandas as pd
hotel= pd.read_csv("hotel_bookings.csv")
hotel.head()
```

```python
#Je více rezervací zrušeno pro městské hotely nebo pro hotely v rezortech?
pd.crosstab(index=hotel['hotel'], columns= hotel['is_canceled'],normalize="index").style.format("{:.2%}")
```

```python
#Pro které kategorie je více zrušených rezervací a pro které naopak více nezrušených?
hotel = hotel[hotel['hotel']=='City Hotel']
bins = [-1,7,30,180,float("inf")]
labels = ['Last-minute','Short-term','Medium-term','Long-term']
hotel['reservation_category']=pd.cut(hotel['lead_time'],bins=bins,labels=labels)
pd.crosstab(index=hotel['reservation_category'], columns= hotel['is_canceled'],normalize='index')
```

```python
#v jaké skupině je celkově nejvíce rezervací?
pd.crosstab(index=hotel['reservation_category'], columns= hotel['is_canceled'])
```

# Halfmaraton

```python
import pandas as pd
hm=pd.read_csv('half_marathon.csv')
hm.head()
```

```python
#Převeď sloupec s časem závodníka na typ datetime
hm['Cas'] = pd.to_datetime(hm['Cas']) 
hm['Cas 2019'] = hm['Cas'].shift(periods = 1)
hm.head()
```

```python
hm = hm[hm['Rok zavodu']==2020]
hm.head()
```

```python
hm['Zmena'] = hm['Cas'] - hm['Cas 2019']
hm.head()
```

```python
hm["Zmena"].mean()
```

```python
# Načte data z URL do DataFrame
data = pd.read_csv(
    "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/datum-cas-shift/excs/swing-states/election-data.csv")
data.head()
```

```python
data = data[(data["state"]=="ALABAMA")&(data["rank"]==1)].sort_values(by = "year")
data.head(4)
```

```python
data['previous_winner_party'] = data["party_simplified"].shift(1)
data.head(4)
```

```python
# Odstraní řádky s chybějícími daty (zejména první řádek, kde je 'previous_winner_party' NaN)
data = data.dropna()
data[data["party_simplified"] != data["previous_winner_party"]]
```
