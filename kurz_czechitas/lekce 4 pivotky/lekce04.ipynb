# 4.lekce PIVOT

```python
import pandas as pd
cholesterol=pd.read_csv("https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/pivot-tabulky/pivot-tabulky/pivot_cholesterol.csv")
cholesterol.head()
```

```python
"""
Obsah cholesterolu	Kategorie
0-20	Low Cholesterol
20-100	Moderate Cholesterol
100-inf	High Cholesterol
"""
bins = [0,20,100,float('inf')]# pozor neni to vcetne bere vetsi nez nula
bins = [-1,20,100,float('inf')]
labels = ['nizky','stredni','vysoky']
cholesterol['category'] = pd.cut(cholesterol["amount"],bins=bins,labels=labels)
cholesterol.head()
```

```python
#pivotka
pd.crosstab(cholesterol["branded_food_category"],cholesterol["category"])
```

```python
#pridavam totals
pd.crosstab(cholesterol["branded_food_category"],cholesterol["category"],margins=True)
```

```python
#chci to v % po radcich
pd.crosstab(cholesterol["branded_food_category"],cholesterol["category"],margins=True,normalize="index")
```

```python
#chci to v % po sloupcich
pd.crosstab(cholesterol["branded_food_category"],cholesterol["category"],margins=True,normalize="columns")
```

```python
#chci to v % na dve desetinne po sloupcich
pd.crosstab(index=cholesterol["branded_food_category"], 
            columns=cholesterol["category"],
            margins=True,
            normalize="columns").style.format("{:.2%}")
```

```python
#novy dataset
cholesterol_fiber = pd.read_csv("https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/pivot-tabulky/pivot-tabulky/pivot_cholesterol_fiber.csv")
cholesterol_fiber.tail()
```

```python
#chci videt prumerne hodnoty
pd.crosstab(index = cholesterol_fiber["branded_food_category"],
            columns= cholesterol_fiber['nutrient_name'],
            values=cholesterol_fiber["amount"],
            aggfunc='mean'
            )
```

# DATE TIME

```python
#tel.signal mame data kdy ztracen a kdy je obnoven
signal_monitoring = pd.read_csv("https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/datum-cas-shift/datum-cas/signal_monitoring.csv")
signal_monitoring.head()
```

```python
signal_monitoring['event_type'].unique()
signal_monitoring.info()
```

```python
# ktery den v tydnu je nejvice vypadku?
# musim pretypovat na datetime
signal_monitoring['event_date_time'] = pd.to_datetime(signal_monitoring['event_date_time'])  
signal_monitoring.info()
```

```python
# prevod na date
signal_monitoring['event_date'] = signal_monitoring['event_date_time'].dt.date
signal_monitoring.head()
```

```python
#den v tydnu
signal_monitoring['event_day_of_week'] = signal_monitoring['event_date_time'].dt.day_name()
signal_monitoring.head()
```

```python
pd.crosstab(index=signal_monitoring['event_day_of_week'],columns=signal_monitoring['event_type'])
```

```python
# ...jak dlouho to trvalo?
signal_monitoring['event_end'] = signal_monitoring['event_date_time'].shift(periods = -1)
signal_monitoring.head()
```

```python
signal_monitoring = signal_monitoring[signal_monitoring["event_type"] == "signal lost"].reset_index(drop=True)
```

```python
# trvani vypadku
signal_monitoring['outage_length'] = signal_monitoring['event_end']-signal_monitoring['event_date_time']
signal_monitoring.head()
```

```python
signal_monitoring.groupby('event_date')['outage_length'].sum()
```
