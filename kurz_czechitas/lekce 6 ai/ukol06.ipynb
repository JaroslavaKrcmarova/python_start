```python
import pandas as pd
import pandas as pd
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
data = pd.read_csv(url)
data.head(5)
```

```python
data.describe()
```

```python
sns.countplot(data, x="species")
```

```python
sns.barplot(data , x="species", y="flipper_length_mm")
```

```python
sns.histplot(data, 
                  x="flipper_length_mm",
                  bins = range(170,240,10),
                  color = "navy"
                  )
```

```python
sns.boxplot(data,
                 x="species", 
                 y="flipper_length_mm",
                 hue ="island"
                 )
```

```python
sns.boxplot(data,
                 x="species", 
                 y="flipper_length_mm",
                 hue ="island",
                 whis=[5,95]
                 )
```
