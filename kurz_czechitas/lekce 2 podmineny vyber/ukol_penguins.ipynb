```python
import pandas as pd
path ="https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
penguins=pd.read_csv(path)
```

```python
#Kolik tučňáků má hmotnost vyšší než 5000 gramů?
penguins5000 = penguins[penguins["body_mass_g"] > 5000]
penguins5000.shape[0]
```

```python
#Které druhy tučňáka mají zastoupení na ostrově Dream?
penguins_dream = penguins[penguins["island"]=='Dream']
penguins_dream["species"].unique()
```

```python
#Kolik tučňáků má hmotnost menší než 3500 gramů a délku křídla (`flipper`) větší než 200 mm
podminka1 = penguins["body_mass_g"]<3500
podminka2 = penguins["flipper_length_mm"]>200
penguins_test = penguins[ (podminka1)
                        & (podminka2)
                        ]
penguins_test.shape
```

```python
#Kolik tučňáků z ostrova Torgersen má ve sloupci `sex` uvedenu hodnotu `female`
podminka1 = penguins["island"]=='Torgersen'
podminka2 = penguins["sex"]=='female'
penguins_test = penguins[ (podminka1)
                        & (podminka2)
                        ]
penguins_test.shape
```
